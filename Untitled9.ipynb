```python
#1.
import random
from statistics import mean, median, mode, harmonic_mean
from collections import Counter
from math import prod

# Generate the list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Mean calculation
def calculate_mean(numbers):
    return mean(numbers)

# (ii) Median calculation
def calculate_median(numbers):
    return median(numbers)

# (iii) Mode calculation
def calculate_mode(numbers):
    freq = Counter(numbers)
    max_freq = max(freq.values())
    modes = [key for key, val in freq.items() if val == max_freq]
    return modes if len(modes) > 1 else modes[0]

# (iv) Weighted mean calculation
def calculate_weighted_mean(values, weights):
    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# (v) Geometric mean calculation
def calculate_geometric_mean(numbers):
    return prod(numbers) ** (1 / len(numbers))

# (vi) Harmonic mean calculation
def calculate_harmonic_mean(numbers):
    return harmonic_mean(numbers)

# (vii) Midrange calculation
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (viii) Trimmed mean calculation
def calculate_trimmed_mean(numbers, trim_percent):
    n = len(numbers)
    trim_count = int(n * trim_percent / 100)
    sorted_numbers = sorted(numbers)[trim_count: n - trim_count]
    return mean(sorted_numbers)

# Calculations for the generated list
mean_value = calculate_mean(int_list)
median_value = calculate_median(int_list)
mode_value = calculate_mode(int_list)
geometric_mean_value = calculate_geometric_mean(int_list)
harmonic_mean_value = calculate_harmonic_mean(int_list)
midrange_value = calculate_midrange(int_list)
trimmed_mean_value = calculate_trimmed_mean(int_list, 10)  # Example with 10% trimming

mean_value, median_value, mode_value, geometric_mean_value, harmonic_mean_value, midrange_value, trimmed_mean_value

# 2.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualizations for the given data

# Frequency & Gaussian Distribution
sns.histplot(int_list2, bins=20, kde=False, color='blue', label='Frequency', stat="density")
sns.kdeplot(int_list2, color='red', label='Gaussian Distribution')
plt.legend()
plt.title("Frequency & Gaussian Distribution")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()

# Frequency Smoothened KDE Plot
sns.histplot(int_list2, bins=20, kde=True, color='green', label='Frequency with KDE')
plt.legend()
plt.title("Frequency with Smoothened KDE Plot")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()

# Gaussian Distribution & Smoothened KDE Plot
sns.kdeplot(int_list2, color='blue', label='Gaussian Distribution')
sns.kdeplot(int_list2, color='orange', label='Smoothened KDE')
plt.legend()
plt.title("Gaussian Distribution & Smoothened KDE Plot")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()


# (ii) Range calculation
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Variance and Standard Deviation calculation
def calculate_variance_and_std(numbers):
    variance = np.var(numbers)
    std_dev = np.std(numbers)
    return variance, std_dev

# (iv) Interquartile Range (IQR) calculation
def calculate_iqr(numbers):
    q1, q3 = np.percentile(numbers, [25, 75])
    return q3 - q1

# (v) Coefficient of Variation calculation
def calculate_coefficient_of_variation(numbers):
    mean_value = np.mean(numbers)
    std_dev = np.std(numbers)
    return (std_dev / mean_value) * 100

# (vi) Mean Absolute Deviation (MAD) calculation
def calculate_mad(numbers):
    mean_value = np.mean(numbers)
    return np.mean([abs(x - mean_value) for x in numbers])

# (vii) Quartile Deviation calculation
def calculate_quartile_deviation(numbers):
    q1, q3 = np.percentile(numbers, [25, 75])
    return (q3 - q1) / 2

# (viii) Range-based Coefficient of Dispersion calculation
def calculate_range_coefficient_of_dispersion(numbers):
    range_value = calculate_range(numbers)
    midrange = (min(numbers) + max(numbers)) / 2
    return (range_value / midrange) * 100

# Calculations for int_list2
range_value = calculate_range(int_list2)
variance, std_dev = calculate_variance_and_std(int_list2)
iqr_value = calculate_iqr(int_list2)
cv_value = calculate_coefficient_of_variation(int_list2)
mad_value = calculate_mad(int_list2)
quartile_deviation = calculate_quartile_deviation(int_list2)
range_dispersion = calculate_range_coefficient_of_dispersion(int_list2)

range_value, variance, std_dev, iqr_value, cv_value, mad_value, quartile_deviation, range_dispersion

# 3.

class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        """
        Initialize the discrete random variable.

        :param outcomes: A list of outcomes (values of the random variable).
        :param probabilities: A list of corresponding probabilities for the outcomes.
        """
        if len(outcomes) != len(probabilities):
            raise ValueError("The number of outcomes must match the number of probabilities.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("The probabilities must sum up to 1.")

        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the random variable.
        """
        return sum(o * p for o, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the random variable.
        """
        mean = self.expected_value()
        return sum(p * ((o - mean) ** 2) for o, p in zip(self.outcomes, self.probabilities))

# Example usage
outcomes = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]

random_variable = DiscreteRandomVariable(outcomes, probabilities)
expected_value = random_variable.expected_value()
variance = random_variable.variance()

expected_value, variance

#4.

import random

# Simulate rolling a six-sided die
def simulate_dice_rolls(num_rolls):
    outcomes = [random.randint(1, 6) for _ in range(num_rolls)]
    return outcomes

# Calculate the expected value and variance
def calculate_expected_value_and_variance(outcomes):
    n = len(outcomes)
    mean_value = sum(outcomes) / n
    variance = sum((x - mean_value) ** 2 for x in outcomes) / n
    return mean_value, variance

# Simulate rolling the die 10,000 times
num_rolls = 10000
dice_outcomes = simulate_dice_rolls(num_rolls)

# Calculate expected value and variance
expected_value, variance = calculate_expected_value_and_variance(dice_outcomes)

expected_value, variance

# 5 .

import numpy as np

def generate_random_samples(distribution, params, sample_size):
    """
    Generate random samples from a specified probability distribution and calculate their mean and variance.

    :param distribution: The name of the distribution ('binomial', 'poisson', etc.).
    :param params: Parameters required for the distribution (e.g., (n, p) for binomial, (lambda,) for Poisson).
    :param sample_size: Number of samples to generate.
    :return: A tuple containing (samples, mean, variance).
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, sample_size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, sample_size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)
    return samples, mean, variance

# Example: Generate samples from a binomial distribution
distribution_name = 'binomial'
distribution_params = (10, 0.5)  # n=10, p=0.5
sample_size = 1000

samples, sample_mean, sample_variance = generate_random_samples(distribution_name, distribution_params, sample_size)

sample_mean, sample_variance

# 6.

import numpy as np

def generate_gaussian_samples(mean, std_dev, sample_size):
    """
    Generate random numbers from a Gaussian distribution and compute statistics.

    :param mean: Mean of the Gaussian distribution.
    :param std_dev: Standard deviation of the Gaussian distribution.
    :param sample_size: Number of random samples to generate.
    :return: A tuple containing (samples, mean, variance, standard deviation).
    """
    samples = np.random.normal(mean, std_dev, sample_size)
    calculated_mean = np.mean(samples)
    variance = np.var(samples)
    standard_deviation = np.std(samples)
    return samples, calculated_mean, variance, standard_deviation

# Example: Generate samples from a Gaussian distribution
gaussian_mean = 0
gaussian_std_dev = 1
sample_size = 1000

samples, sample_mean, sample_variance, sample_std_dev = generate_gaussian_samples(
    gaussian_mean, gaussian_std_dev, sample_size
)

sample_mean, sample_variance, sample_std_dev

#7.

# Importing required libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, pearsonr

# Load the dataset
tips = sns.load_dataset('tips')

# (i) Function to calculate skewness
def calculate_skewness(column):
    return skew(column)

# (ii) Program to determine skewness type
def skewness_type(column):
    skewness = calculate_skewness(column)
    if skewness > 0:
        return "Positive Skewness"
    elif skewness < 0:
        return "Negative Skewness"
    else:
        return "Approximately Symmetric"

# (iii) Function to calculate covariance between two columns
def calculate_covariance(column1, column2):
    return np.cov(column1, column2)[0, 1]

# (iv) Function to calculate Pearson correlation coefficient
def calculate_pearson_correlation(column1, column2):
    return pearsonr(column1, column2)[0]

# (v) Visualization of correlation using scatter plot
def visualize_correlation(column1, column2):
    plt.figure(figsize=(8, 6))
    plt.scatter(column1, column2, alpha=0.7, edgecolor='k')
    plt.title('Scatter Plot of Two Columns', fontsize=16)
    plt.xlabel('Total Bill', fontsize=14)
    plt.ylabel('Tip', fontsize=14)
    plt.grid(True)
    plt.show()

# Running calculations
total_bill = tips['total_bill']
tip = tips['tip']

# (i) Skewness
skewness_total_bill = calculate_skewness(total_bill)
skewness_tip = calculate_skewness(tip)

# (ii) Skewness type
skewness_type_total_bill = skewness_type(total_bill)
skewness_type_tip = skewness_type(tip)

# (iii) Covariance
covariance = calculate_covariance(total_bill, tip)

# (iv) Pearson correlation coefficient
pearson_correlation = calculate_pearson_correlation(total_bill, tip)

# (v) Visualization
visualize_correlation(total_bill, tip)

(skewness_total_bill, skewness_tip, skewness_type_total_bill, skewness_type_tip, covariance, pearson_correlation)

# 8.

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) for a normal distribution.

    Parameters:
        x (float): The value of the random variable.
        mean (float): The mean () of the normal distribution.
        std_dev (float): The standard deviation () of the normal distribution.

    Returns:
        float: The PDF value for the given x.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    return coefficient * math.exp(exponent)

# Example usage:
x_value = 5
mean_value = 4
std_dev_value = 1.5

pdf_value = normal_pdf(x_value, mean_value, std_dev_value)
print(f"The PDF value for x={x_value} in the normal distribution with mean={mean_value} and std_dev={std_dev_value} is {pdf_value:.6f}")


#9.

import math

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) for an exponential distribution.

    Parameters:
        x (float): The value of the random variable (x >= 0).
        rate (float): The rate parameter () of the exponential distribution ( > 0).

    Returns:
        float: The CDF value for the given x.
    """
    if x < 0:
        return 0.0  # CDF is 0 for x < 0 in an exponential distribution
    return 1 - math.exp(-rate * x)

# Example usage:
x_value = 2
rate_value = 0.5

cdf_value = exponential_cdf(x_value, rate_value)
print(f"The CDF value for x={x_value} with rate parameter ={rate_value} is {cdf_value:.6f}")

# 10.
import math

def poisson_pmf(k, rate):
    """
    Calculate the probability mass function (PMF) for a Poisson distribution.

    Parameters:
        k (int): The number of events (k).
        rate (float): The rate parameter () of the Poisson distribution.

    Returns:
        float: The PMF value for the given k.
    """
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")

    # Poisson PMF formula
    return (rate ** k * math.exp(-rate)) / math.factorial(k)

# Example usage:
k_value = 3
rate_value = 2.5

pmf_value = poisson_pmf(k_value, rate_value)
print(f"The PMF value for k={k_value} in the Poisson distribution with rate={rate_value} is {pmf_value:.6f}")


# 11.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Conversion rates
successes = [np.sum(old_layout), np.sum(new_layout)]  # Number of purchases
n_trials = [len(old_layout), len(new_layout)]  # Number of visitors

# Perform z-test
z_stat, p_value = proportions_ztest(successes, n_trials)

z_stat, p_value

# 12

from statsmodels.stats.weightstats import ztest

# Generate sample data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Perform z-test on paired data (differences before and after)
differences = after_program - before_program
z_stat, p_value = ztest(differences, value=0)

z_stat, p_value

# 13.

# Generate sample data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Perform z-test on paired data (differences before and after)
differences = before_drug - after_drug
z_stat, p_value = ztest(differences, value=0)

z_stat, p_value


# 14

# Generate sample data
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Perform one-sample z-test (mean < 5)
z_stat, p_value = ztest(response_times, value=5, alternative='smaller')

z_stat, p_value

# 15

from scipy.stats import ttest_ind

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform independent t-test
t_stat, p_value = ttest_ind(layout_a_clicks, layout_b_clicks, equal_var=False)

# Degrees of freedom calculation
n_a = len(layout_a_clicks)
n_b = len(layout_b_clicks)
var_a = np.var(layout_a_clicks, ddof=1)
var_b = np.var(layout_b_clicks, ddof=1)
degrees_of_freedom = ((var_a / n_a + var_b / n_b) ** 2) / \
    ((var_a / n_a) ** 2 / (n_a - 1) + (var_b / n_b) ** 2 / (n_b - 1))

t_stat, degrees_of_freedom, p_value

# 16

# Data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform independent t-test
t_stat, p_value = ttest_ind(existing_drug_levels, new_drug_levels, equal_var=False)

t_stat, p_value

# 17

from scipy.stats import ttest_rel

# Data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform paired t-test
t_stat, p_value = ttest_rel(pre_intervention_scores, post_intervention_scores)

t_stat, p_value

# 18

# Generate synthetic salary data
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform independent t-test
t_stat, p_value = ttest_ind(male_salaries, female_salaries, equal_var=False)

t_stat, p_value

# 19

# Data
version1_scores = [85, 88, 82, 89, 87, 84, 90, 88, 85, 86, 91, 83, 87, 84, 89, 86, 84, 88, 85, 86, 89, 90, 87, 88, 85]
version2_scores = [80, 78, 83, 81, 79, 82, 76, 80, 78, 81, 77, 82, 80, 79, 82, 79, 80, 81, 79, 82, 79, 78, 80, 81, 82]

# Perform independent t-test
t_stat, p_value = ttest_ind(version1_scores, version2_scores, equal_var=False)

t_stat, p_value

# 20

import numpy as np
from scipy import stats

# Scores for each branch
branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

# Convert the scores to numpy arrays for easier calculations
a = np.array(branch_a_scores)
b = np.array(branch_b_scores)

# Calculate means
mean_a = np.mean(a)
mean_b = np.mean(b)

# Calculate standard deviations
std_a = np.std(a, ddof=1)  # sample standard deviation
std_b = np.std(b, ddof=1)

# Calculate sample sizes
n_a = len(a)
n_b = len(b)

# Calculate the t-statistic
t_statistic = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))

# Calculate degrees of freedom
df = n_a + n_b - 2

# Calculate the p-value
p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # two-tailed test

# Display results
print(f"Branch A Mean: {mean_a}")
print(f"Branch B Mean: {mean_b}")
print(f"T-Statistic: {t_statistic}")
print(f"Degrees of Freedom: {df}")
print(f"P-Value: {p_value}")

# Set a significance level
alpha = 0.05
if p_value < alpha:
    print("There is a statistically significant difference between the customer satisfaction scores of the two branches.")
else:
    print("There is no statistically significant difference between the customer satisfaction scores of the two branches.")


# 21

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate sample data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a contingency table
data = pd.crosstab(age_groups, voter_preferences)

# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Output the results
chi2_stat, p_value, dof, expected

# 22

import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

chi2_stat, p_value, dof, expected

# 23

# Sample data: Job performance levels before (rows) and after (columns) training
data = np.array([[50, 30, 20], [30, 40, 30], [20, 30, 40]])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

chi2_stat, p_value, dof, expected


# 24

from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform ANOVA test
f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

f_stat, p_value
















```
